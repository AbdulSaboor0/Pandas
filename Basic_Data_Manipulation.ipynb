```python
### It is not always a good idea to replace the missing values with zero,
### if the data is less, also if in case the data may contain string values or if we calculate the mean.
### Hence, we fill specific values for specific column.


df2 = df.fillna({
    'Type 2' : 'No Data',
    'HP': 0,
    'Attack'	: 0,
    'Sp. Atk': 0,
    'Sp. Def': 0,
    'Speed': 0,
    'Defense': 0

})
df2
```

```python
### Droping unwanted Columns

df2 = df2.drop('#', axis =1)
df2
```

```python

df2.isnull()

```

### ***Changing Zero values to more logical values***

```python
### Carry forwarding the data from previous row or we can also use bfill

df2 = df.fillna(method = 'ffill')
df2
```

```python
### Linear Interpolate

df2 = df.interpolate()
df2

### We also have other option like quadratic, cubic interpolate
```
